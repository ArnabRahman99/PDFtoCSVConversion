```python
# importing libraries

# Note for camelot library: dependencies are tkinter and ghostscript: check the camelot documentation for installations
import camelot
import pandas as pd
import numpy 
import time
from pathlib import Path

def pdf_converter():
    
    """
    
    This is a function which takes the name of a pdf and converts tables into a csv file.
    
    Input: Name of pdf (including the pdf extension)
    Output: Csv file with tables from the pdf
    
    Suggested Workflow: Split PDFs into schedules, then run this script to get excel files.
    
    """
    
    # getting name of pdf
    pdf_title = str(input('Enter the title of the split pdf you would like to convert to a csv'))
    my_file = Path(pdf_title)

    # checking if pdf exists
    if my_file.is_file():
        
        # add if statement for streaming option
        weird_or_no = str(input('Are your pdf tables well structured? y or n'))
        
        if weird_or_no == 'y':
            
            # success case for well structured tables, don't need camelot streaming
            
            print('file exists... reading well structured table')

            # parsing pdf pages and converting to tables
            pdf = camelot.read_pdf(pdf_title, pages = 'all')
            print ('pages have been read... concatenating now...')

            # setting a dataframe to append to
            dfnew = pdf[0].df

            # setting up iteration parameters
            num_of_table_index = len(list(pdf))

            # iterating and appending 
            for i in range (num_of_table_index):
                dftemp = pdf[i].df
                dfnew = dfnew.append(dftemp)

            # cleaning final dataframe
            df_final = dfnew.drop_duplicates()

            # creating a new csv file with info
            df_final.to_csv(pdf_title.split('.pdf')[0]+'.csv')
            print ('table has been converted in the directory')
            
        elif weird_or_no == 'n':
            
            # success case for un-structured tables, need camelot streaming and imaging, probably not the best option
            print ('file exists... reading un-structured table..')
            
            # parsing pdf pages and converting to tables
            pdf = camelot.read_pdf(pdf_title, flavor='stream', pages = 'all')
            print ('pages have been read... concatenating now...')

            # setting a dataframe to append to
            dfnew = pdf[0].df

            # setting up iteration parameters
            num_of_table_index = len(list(pdf))

            # iterating and appending 
            for i in range (num_of_table_index):
                dftemp = pdf[i].df
                dfnew = dfnew.append(dftemp)

            # cleaning final dataframe
            df_final = dfnew.drop_duplicates()

            # creating a new csv file with info
            df_final.to_csv(pdf_title.split('.pdf')[0]+'.csv')
            print ('table has been converted in the directory')
        
        
        else: 
            print ('y or n was not chosen')
                  

    else:
        print('file does not exist')
    
# main method
if __name__ == '__main__':
    
    # time calculation
    start_time = time.time()
    
    # pdf conversion function 
    pdf_converter()

print("--- %s seconds ---" % (time.time() - start_time))

```
